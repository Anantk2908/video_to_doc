```python
from datetime import timedelta

def format_time(seconds):
    return str(timedelta(seconds=round(seconds, 2)))

# Your updated loop
data = ""
for i in range(len(result["segments"])):
    text = result["segments"][i]["text"]
    start_time = format_time(result["segments"][i]['start'])
    end_time = format_time(result["segments"][i]['end'])
    
    data += f'{text} Start Time: {start_time} End Time: {end_time}\n'

print(data)
```

```python
prompt = f"""You are an expert in creating detailed Word documents from video transcripts. Structure the document to include a Table of Contents  and an Introduction as the first sections. Use logical headings throughout, based on the transcript content. Include relevant quotes and timestamps (e.g., [00:05:12]) for clarity. Skip any sponsored segments and focus only on the main content of the video.
Video Transcript: {data}
Document Title:
Word Document:
"""
```

```python
prompt = (
    "You are an expert in creating structured JSON responses from video transcripts to be converted into detailed Word documents. "
    "Format the JSON to include a 'title' field, followed by a 'table_of_contents' field as an array listing the main sections and subsections. "
    "Then use a 'sections' field as an array, where each object includes:\n"
    "1. A 'heading' field for the main section title,\n"
    "2. A 'subsections' field, which is an array of objects. Each object in 'subsections' should include a 'subheading' field and a 'content' field with relevant quotes and timestamps (e.g., [00:05:12]) based on the transcript content. Skip any sponsored segments and focus only on the main content of the video.\n\n"
    "Format Example:\n"
    "{\n"
    "    'title': 'Document Title',\n"
    "    'table_of_contents': ['Section 1', 'Section 1.1', 'Section 2', 'Section 2.1'],\n"
    "   as 'sections': [\n"
    "        {\n"
    "            'heading': 'Section 1',\n"
    "            'subsections': [\n"
    "                {\n"
    "                    'subheading': 'Section 1.1',\n"
    "                    'content': '...'\n"
    "                }\n"
    "            ]\n"
    "        }\n"
    "    ]\n"
    "}\n\n"
    f"Video Transcript: {data}\n"
    "JSON Output:"
)
```

```python
print(prompt)
```

```python
import google.generativeai as genai
import os

genai.configure(api_key=os.environ["GEMINI_API_KEY"])
model = genai.GenerativeModel(model_name="gemini-1.5-pro-002")
print("total_tokens: ", model.count_tokens(prompt))
response = model.generate_content(prompt)
print(response.text)
```

```python
from docx import Document
from bs4 import BeautifulSoup
import markdown
import re

input_text = response.text

# Step 1: Extract the title for the filename
title_match = re.search(r'^##\s*(.+)', input_text, re.MULTILINE)
if title_match:
    title = title_match.group(1).strip()
    filename = re.sub(r'[^\w\s-]', '', title) + '.docx'  # Remove special characters from title for filename
else:
    filename = 'Document.docx'  # Default filename if title is not found

# Step 2: Convert Markdown to HTML
html = markdown.markdown(input_text)

# Step 3: Parse HTML using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Step 4: Create a new Word Document
doc = Document()

# Function to add elements to the document
def add_element(element):
    if element.name == 'h1':
        doc.add_heading(element.get_text(), level=1)
    elif element.name == 'h2':
        doc.add_heading(element.get_text(), level=2)
    elif element.name == 'h3':
        doc.add_heading(element.get_text(), level=3)
    elif element.name == 'h4':
        doc.add_heading(element.get_text(), level=4)
    elif element.name == 'p':
        # Check for bold or italic text
        paragraph = doc.add_paragraph()
        for content in element.contents:
            if content.name == 'strong':
                run = paragraph.add_run(content.get_text())
                run.bold = True
            elif content.name == 'em':
                run = paragraph.add_run(content.get_text())
                run.italic = True
            else:
                run = paragraph.add_run(str(content))
    elif element.name == 'ul':
        for li in element.find_all('li'):
            doc.add_paragraph(li.get_text(), style='List Bullet')
    elif element.name == 'ol':
        for li in element.find_all('li'):
            doc.add_paragraph(li.get_text(), style='List Number')
    elif element.name == 'blockquote':
        paragraph = doc.add_paragraph(element.get_text(), style='Intense Quote')
    else:
        doc.add_paragraph(element.get_text())

# Step 5: Iterate over the parsed HTML and add to the document
for element in soup:
    if hasattr(element, 'name'):
        add_element(element)

# Step 6: Save the document with the dynamic filename
doc.save(filename)
print(f"Document saved as: {filename}")
```
